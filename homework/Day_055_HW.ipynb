- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
%matplotlib inline
from sklearn import datasets
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(17)
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(X.shape)
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
def KM_clustering(n_clusters, n_init=10, init='k-means++'):
    #init{‘k-means++’, ‘random’, ndarray, callable}, default=’k-means++’
    est = KMeans(n_clusters=n_clusters, n_init=n_init, init=init)
    fig = plt.figure(figsize=(8, 6))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

    ## fit data
    est.fit(X)
    labels = est.labels_
    ax.scatter(X[:, 3], X[:, 0], X[:, 2],
               c=labels.astype(np.float), s=80, edgecolor='k')

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    if init == None:
        ax.set_title('Clusters: %d, n_init: %d' %(n_clusters, n_init))
    else:
        ax.set_title('Clusters: {}, n_init: {}, init: {}'.format(n_clusters, n_init, init))
    ax.dist = 12
    plt.show()
```

```python
KM_clustering(3)
```

```python
KM_clustering(5)
```

```python
KM_clustering(3, n_init=7, init='k-means++')
```

```python
KM_clustering(3, n_init=7, init='random')
```

```python
KM_clustering(3, n_init=1, init='random')
```

```python
KM_clustering(7, n_init=1, init='random')
```

```python
KM_clustering(7, n_init=10, init='random')
```
