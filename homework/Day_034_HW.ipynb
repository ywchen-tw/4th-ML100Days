## 練習時間
假設我們資料中類別的數量並不均衡，在評估準確率時可能會有所偏頗，試著切分出 y_test 中，0 類別與 1 類別的數量是一樣的 (亦即 y_test 的類別是均衡的)

```python
import numpy as np
X = np.arange(1000).reshape(200, 5)
y = np.zeros(200)
y[:40] = 1
```

```python
y
```

可以看見 y 類別中，有 160 個 類別 0，40 個 類別 1 ，請試著使用 train_test_split 函數，切分出 y_test 中能各有 10 筆類別 0 與 10 筆類別 1 。(HINT: 參考函數中的 test_size，可針對不同類別各自作切分後再合併)

```python
from sklearn.model_selection import train_test_split
X_0, y_0 = X[y==0], y[y==0]
X_1, y_1 = X[y==1], y[y==1]
```

```python
X_train_0, X_test_0, y_train_0, y_test_0 = train_test_split(X_0, y_0, test_size=10/len(y_0), random_state=42)
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=10/len(y_1), random_state=42)
```

```python
X_train = np.concatenate((X_train_0, X_train_1), axis=None)
X_train
```

```python
y_train = np.concatenate((y_train_0, y_train_1), axis=None)
y_train
```

```python
X_test = np.concatenate((X_test_0, X_test_1), axis=None)
X_test
```

```python
y_test = np.concatenate((y_test_0, y_test_1), axis=None)
y_test
```
